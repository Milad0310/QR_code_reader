```python
import cv2
from pyzbar.pyzbar import decode
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/barcode_qrcode.jpg')
decode_img =decode(img)
```

```python
(decode_img)
```

```python
for d in decode_img:
    cv2.rectangle(img, (d.rect.left, d.rect.top), (d.rect.left + d.rect.width, d.rect.top + d.rect.height),(255,0,0), 2)
    cv2.polylines(img,[np.array(d.polygon)], True, (0, 255, 0), 2)
```

```python
  cv2_imshow(img)
```
